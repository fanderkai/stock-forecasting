## Data Wrangling

### Convert prices to float

```python
price_cols = ['Close/Last', 'Open', 'High', 'Low']
for col in price_cols:
  stock_df[col] = stock_df[col].apply(lambda x: float(str(x).replace("$", "")))
```

### Convert date to datetime

```python
stock_df['Date'] = stock_df["Date"].apply(lambda x: x.replace("-", "/"))
stock_df['Date'] = pd.to_datetime(stock_df['Date'], format="%m/%d/%Y")
```

### Rename and sort

```python
stock_df = stock_df.rename(columns={"Close/Last": "Close"})
```

```python
stock_df = stock_df.sort_values(["Company", "Date"]).reset_index(drop=True)
```
