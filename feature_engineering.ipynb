```python
import pandas as pd
```

```python
stock_df = pd.read_csv(
    "/content/stock_cleaned.csv",
    parse_dates=["Date"]
)

stock_df.head()
```

```python
for lag in [1, 5, 21]:
    stock_df[f"lag_{lag}"] = stock_df["Close"].shift(lag)

for span in [12, 26]:
    stock_df[f"ema_{span}"] = stock_df["Close"].ewm(span=span, adjust=False).mean()

stock_df["macd"] = stock_df["ema_12"] - stock_df["ema_26"]
stock_df["momentum_10"]   = stock_df["Close"].pct_change(10)
stock_df["volatility_21"] = stock_df["Close"].rolling(21).std()

stock_df["volume_change"]  = stock_df["Volume"].pct_change()
stock_df["volume_ma_7"]    = stock_df["Volume"].rolling(7).mean()
stock_df["relative_volume"] = stock_df["Volume"] / stock_df["volume_ma_7"]

stock_df["daily_range"] = (stock_df["High"] - stock_df["Low"]) / stock_df["Close"]
stock_df["oc_return"] = (stock_df["Close"] - stock_df["Open"]) / stock_df["Open"]
```
